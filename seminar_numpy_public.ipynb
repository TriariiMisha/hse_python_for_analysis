````python
a = np.array([[1, 0, 1], [2, 0, 2], [3, 2, 3], [4, 5, 4]])
a
````

### Поделаем градиент
Сделать функцию, которая получает на вход:

- пиксель цвета (начальный);
- пиксель цвета (конечный);
- количество цветов.

Она должна рисовать квадрат размера 1024 на 1024 с градиентом.

````python
import numpy as np
import matplotlib.pyplot as plt


SIZE = 1024


def gradient(start_color: np.ndarray, end_color: np.ndarray, n_colors: int = 2):
    n_repeat = SIZE // n_colors
    
    square = np.repeat(
        np.expand_dims(
            np.repeat(
                np.linspace(
                    start_color, end_color, n_colors
                ).astype(int), n_repeat, axis=0
            ), axis=0
        ), SIZE, axis=0
    )

    plt.imshow(square)
````

````python
gradient(
    [0, 0, 0],
    [255, 255, 255],
    200
)
````
